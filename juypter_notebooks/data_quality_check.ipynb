```python
import duckdb as ddb
import pandas as pd
```

```python
# set up connection
con = ddb.connect("../air_quality.db")
con
```

```python
# looking for data with specific parameter
df = con.query("select * from raw.air_quality where parameter in ('so2', 'pm10', 'pm25')").to_df()
df.head()
```
